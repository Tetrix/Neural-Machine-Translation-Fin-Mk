```python
import pandas as pd
pd.options.mode.chained_assignment = None 
import numpy as np
import re
import nltk
from sklearn.manifold import TSNE
import numpy as np

from gensim.models import word2vec, KeyedVectors

from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
# load mk embeddings
mk_embeddings = KeyedVectors.load('../data/embeddings/mk_embeddings.bin')
```

```python
# load fi embeddings
fi_embeddings = KeyedVectors.load('../data/embeddings/fi_embeddings.bin')
```

```python
keys_fin = ['mies', 'nainen', 'englanti', 'vuotta', 'kaunis', 'vuori', 'puhua', 'missä', 'peikko']
keys_mk = ['маж','жена', 'англија', 'година', 'убава', 'планина', 'зборува', 'каде', 'трол']

def get_clusters(keys, model):
    embedding_clusters = []
    word_clusters = []
    for word in keys:
        embeddings = []
        words = []
        for similar_word, _ in model.wv.most_similar(word, topn=20):
            words.append(similar_word)
            embeddings.append(model.wv[similar_word])
        embedding_clusters.append(embeddings)
        word_clusters.append(words)
    return embedding_clusters, word_clusters

embedding_clusters_fin, word_clusters_fin = get_clusters(keys_fin, fi_embeddings)
embedding_clusters_mk, word_clusters_mk = get_clusters(keys_mk, mk_embeddings)
```

```python
embedding_clusters_fin = np.array(embedding_clusters_fin)
embedding_clusters_mk = np.array(embedding_clusters_mk)

n, m, k = embedding_clusters_fin.shape
n, m, k = embedding_clusters_mk.shape

tsne_model_fin_2d = TSNE(perplexity=20, n_components=2, init='pca', n_iter=2000, random_state=32)
tsne_model_mk_2d = TSNE(perplexity=20, n_components=2, init='pca', n_iter=2000, random_state=32)

embeddings_fin_2d = np.array(tsne_model_fin_2d.fit_transform(embedding_clusters_fin.reshape(n * m, k))).reshape(n, m, 2)
embeddings_mk_2d = np.array(tsne_model_mk_2d.fit_transform(embedding_clusters_mk.reshape(n * m, k))).reshape(n, m, 2)
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm


def tsne_plot_similar_words(title, labels, embedding_clusters, word_clusters, alpha, filename=None):
    plt.figure(figsize=(16, 9))
    colors = cm.rainbow(np.linspace(0, 1, len(labels)))
    for label, embeddings, words, color in zip(labels, embedding_clusters, word_clusters, colors):
        x = embeddings[:, 0]
        y = embeddings[:, 1]
        plt.scatter(x, y, c=[color], alpha=alpha, label=label)
        for i, word in enumerate(words):
            plt.annotate(word, alpha=0.5, xy=(x[i], y[i]), xytext=(5, 2),
                         textcoords='offset points', ha='right', va='bottom', size=8)
    plt.legend(loc=4)
    plt.title(title)
    plt.grid(True)
    if filename:
        plt.savefig(filename, format='eps')
    plt.show()
```

```python
tsne_plot_similar_words('Word Clusters Finnish', keys_fin, embeddings_fin_2d, word_clusters_fin, 0.7,
                        'fin_clusters.eps')
```

```python
tsne_plot_similar_words('Word Clusters Macedonian', keys_mk, embeddings_mk_2d, word_clusters_mk, 0.7,
                        'mk_clusters.eps')
```
